```python
def safe_to_place(chessboard, r, c, size):
    for i in range(r):
        if chessboard[i][c] == 1:
            return False

    i, j = r - 1, c - 1
    while i >= 0 and j >= 0:
        if chessboard[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i, j = r - 1, c + 1
    while i >= 0 and j < size:
        if chessboard[i][j] == 1:
            return False
        i -= 1
        j += 1

    return True

def n_queens_solver_util(chessboard, r, size, result_list):
    if r == size:
        result_list.append(["".join("Q" if cell == 1 else "." for cell in row) for row in chessboard])
        return

    for c in range(size):
        if safe_to_place(chessboard, r, c, size):
            chessboard[r][c] = 1
            n_queens_solver_util(chessboard, r + 1, size, result_list)
            chessboard[r][c] = 0

def solve_n_queens_problem(size):
    board = [[0]*size for _ in range(size)]
    solutions = []
    n_queens_solver_util(board, 0, size, solutions)
    return solutions

N = 8
all_solutions = solve_n_queens_problem(N)
print(f"Total {N}-Queens solutions: {len(all_solutions)}\n")
for sol in all_solutions:
    for row in sol:
        print(row)
    print()
```
