```python

jug_a_capacity = 4
jug_b_capacity = 3
target_amount = 2

visited_states = set()

def dfs_jug(state_a, state_b, steps):

    if (state_a, state_b) in visited_states:
        return False

    visited_states.add((state_a, state_b))
    steps.append((state_a, state_b))

    if state_a == target_amount:
        print("Solution Achieved\n")
        print("Step Sequence:")
        for s in steps:
            print(s)
        return True

    if state_a < jug_a_capacity:
        print("Action: Fill Jug A")
        if dfs_jug(jug_a_capacity, state_b, steps.copy()):
            return True

    if state_b < jug_b_capacity:
        print("Action: Fill Jug B")
        if dfs_jug(state_a, jug_b_capacity, steps.copy()):
            return True

    if state_a > 0:
        print("Action: Empty Jug A")
        if dfs_jug(0, state_b, steps.copy()):
            return True

    if state_b > 0:
        print("Action: Empty Jug B")
        if dfs_jug(state_a, 0, steps.copy()):
            return True

    if state_a > 0 and state_b < jug_b_capacity:
        pour_amount = min(state_a, jug_b_capacity - state_b)
        print("Action: Pour Jug A -> Jug B")
        if dfs_jug(state_a - pour_amount, state_b + pour_amount, steps.copy()):
            return True

    if state_b > 0 and state_a < jug_a_capacity:
        pour_amount = min(state_b, jug_a_capacity - state_a)
        print("Action: Pour Jug B -> Jug A")
        if dfs_jug(state_a + pour_amount, state_b - pour_amount, steps.copy()):
            return True

    return False

dfs_jug(0, 0, [])
```
